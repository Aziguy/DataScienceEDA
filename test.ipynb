```python
import pandas as pd
```

```python
df = pd.read_csv('./datas/googleplaystore.csv')
df.shape
# 10841 lignes et 13 colonnes dans le fichier
```

```python
# VÃ©rifie si des applications sont dupliquÃ©es (ont strictement le mÃªme nom)
df[df['App'].duplicated()]
```

```python
# Supprime ces lignes
df = df.drop_duplicates(subset='App')
```

```python
df.shape
# On a donc maintenant 9660 lignes au lieu de 10841 lignes
```

```python
# VÃ©rifions si des valeurs sont manquantes
df.isna().sum()

# La plupart des valeurs manquantes sont dans la colonne rating
```

```python
df.loc[df['Rating'].isna(), :]
```

```python
# OÃ¹ sont les donnÃ©es manquantes ? 
df.isnull().sum()
```

```python
# Affichage des quelques premiÃ¨res lignes oÃ¹ il manque le rating
df[df.Rating.isnull()].head()
```

```python
df = df.dropna(subset=['Rating'])
```

```python
df.shape
```

```python
df['Rating'].describe()
# On s'aperÃ§oit qu'au moins une valeur n'est pas comprise entre 1 et 5
```

```python
df = df[(df['Rating'] >= 1) & (df['Rating'] <= 5)]
df.describe()
```

```python
df.isna().sum()
```

```python
df['Rating'].astype(float)
```

```python
df['Category'] = [el.replace('_', ' ').capitalize() for el in df['Category']]
```

```python
# Nettoyage des unitÃ©s (kilos et mÃ©ga octets)

df['Size'] = df['Size'].apply(lambda x: float(str(x).replace('k', ''))/1024 if 'k' in str(x) else x)
df['Size'] = df['Size'].apply(lambda x: float(str(x).replace('M', '')) if 'M' in str(x) else x)
df['Size'] = df['Size'].apply(lambda x: float(str(x).replace('Varies with device', 'NaN')) if 'Varies with device' in str(x) else x)
df['Size']
```

```python
df['Installs'] = df['Installs'].apply(lambda x: float(str(x).replace(',', '').replace('+', '')) if '+' in str(x) else x)
df['Installs']
```

```python
df.head()
```

```python
df.to_csv('./datas/clean.csv')
```
