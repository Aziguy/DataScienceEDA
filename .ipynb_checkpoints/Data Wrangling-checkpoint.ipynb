```python
import pandas as pd
```

```python
df = pd.read_csv('./datas/googleplaystore.csv')
df.shape
# 10841 lignes et 13 colonnes dans le fichier
```

```python
# Vérifie si des applications sont dupliquées (ont strictement le même nom)
df[df['App'].duplicated()]
```

```python
# Supprime ces lignes
df = df.drop_duplicates(subset='App')
```

```python
df.shape
# On a donc maintenant 9660 lignes au lieu de 10841 lignes
```

```python
# Vérifions si des valeurs sont manquantes
df.isna().sum()

# La plupart des valeurs manquantes sont dans la colonne rating
```

```python
df.loc[df['Rating'].isna(), :]
```

```python
# Où sont les données manquantes ? 
df.isnull().sum()
```

```python
# Affichage des quelques premières lignes où il manque le rating
df[df.Rating.isnull()].head()
```

```python
df = df.dropna(subset=['Rating'])
```

```python
df.shape
```

```python
df['Rating'].describe()
# On s'aperçoit qu'au moins une valeur n'est pas comprise entre 1 et 5
```

```python
df = df[(df['Rating'] >= 1) & (df['Rating'] <= 5)]
df.describe()
```
